```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# Carrega o dataset com o delimitador correto
file_path = 'PEDE_PASSOS_DATASET_FIAP.csv'
df = pd.read_csv(file_path, delimiter=';')

# Remove as linhas com valores nulos
df_cleaned = df.dropna()

# Realiza a contagem de valores em cada coluna 'PEDRA'
count_2020 = df_cleaned['PEDRA_2020'].value_counts()
count_2021 = df_cleaned['PEDRA_2021'].value_counts()
count_2022 = df_cleaned['PEDRA_2022'].value_counts()

# Ordenar as contagens pelas categorias específicas: Topázio, Ametista, Ágata e Quartzo
ordered_categories = ['Topázio', 'Ametista', 'Ágata', 'Quartzo']

# Reindexar as contagens com base na ordem desejada
count_2020 = count_2020.reindex(ordered_categories)
count_2021 = count_2021.reindex(ordered_categories)
count_2022 = count_2022.reindex(ordered_categories)

# Criar o gráfico de barras com rótulos de dados
plt.figure(figsize=(12, 6))

# Gráfico para 2020
plt.subplot(1, 3, 1)
bars_2020 = count_2020.plot(kind='bar')
plt.title('Contagem de PEDRA 2020')
plt.xlabel('Valores')
plt.ylabel('Frequência')
# Adicionar rótulos de dados nas barras
for bar in bars_2020.patches:
    bars_2020.annotate(format(bar.get_height(), '.0f'),
                       (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                       ha='center', va='bottom')

# Gráfico para 2021
plt.subplot(1, 3, 2)
bars_2021 = count_2021.plot(kind='bar')
plt.title('Contagem de PEDRA 2021')
plt.xlabel('Valores')
# Adicionar rótulos de dados nas barras
for bar in bars_2021.patches:
    bars_2021.annotate(format(bar.get_height(), '.0f'),
                       (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                       ha='center', va='bottom')

# Gráfico para 2022
plt.subplot(1, 3, 3)
bars_2022 = count_2022.plot(kind='bar')
plt.title('Contagem de PEDRA 2022')
plt.xlabel('Valores')
# Adicionar rótulos de dados nas barras
for bar in bars_2022.patches:
    bars_2022.annotate(format(bar.get_height(), '.0f'),
                       (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                       ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
